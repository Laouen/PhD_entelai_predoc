```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
def read_csv(schema):
    features_path = f'../{schema}/execution_state/feature_importance/output/'
    df = pd.DataFrame({
        'features': np.load(f'../{schema}/execution_state/data_preprocess/output/features.npy'),
        'importances_mean': np.load(f'{features_path}/importances_mean.npy'),
        'importances_std': np.load(f'{features_path}/importances_std.npy')
    })

    df['feature_categories'] = df['features'].apply(lambda feature: feature.split('__')[0])
    return df
```

```python
df = read_csv('Migra√±as_vs_otras')
```

```python
sns.barplot(x='feature_categories', y='importances_mean', data=df)
plt.title('Feature importance mean')
plt.xlabel('Features')
plt.ylabel('Importance mean')
plt.show()
```
